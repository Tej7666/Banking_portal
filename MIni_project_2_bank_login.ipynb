```python
db= {}
list1 = ['Username','Password','Balance']

def title ():
    print ('-'*50)
    print ('Reserve Bank of India'.center(50),'\n')
    print ('Login Page :')
    print ('1 : Create login for New User ')
    print ('2 : Existing user Login')
    print ('-'*50)
    
def add_details():
    ano = int(input('Enter account number:'))
    for item in list1:
        list2.append(input('Enter %s:'%item))
        
    db[ano] = dict(zip(list1,list2))
    print('Account Created ')
    print ('-'*50)
    return


def check_acc_no(ano):
    if ano not in db:
        raise NameError('User not found please create new account !!!')
    return True
        
def check_balance(ano):
    if check_acc_no(ano):
        print(f"Balance is now {db[ano]['Balance']}")

def withdraw(ano,amount):
    if check_acc_no(ano):
        if int(db[ano]['Balance']) < amount:
            raise MemoryError('Balance is Low trasancation cant be processed)
        else:
            db[ano]['Balance'] = int(db[ano]['Balance']) - amount
            check_balance(ano)
            
def deposit(ano,amount): 
    if check_acc_no(ano):
        db[ano]['Balance'] = int(db[ano]['Balance']) + amount
        check_balance(ano)
    else :
        if amount < 0:
            raise ImportError('Please enter valid amount')

def choices():
    print('1 - Check Balance')
    print('2 - Withdraw')
    print('3 - Deposit')
    print('4 - Exit')
    op = input('Enter your Choice:')
    op = int(op)
    if  op == 1:
        check_balance(ano)
        
    elif op == 2:
        amt = int(input('Enter the amount to withdraw : '))
        try:
            withdraw(ano,amt)
        except MemoryError as err:
            print('-'*30)
            print(err)
            print('-'*30)
    elif op == 3:
        amt = int(input('Enter the amount to deposit : '))
        try:
            deposit(ano,amt)
        except ImportError as ier:
            print('-'*30)
            print(ier)
            print('-'*30)  
            

while True :
    list2 = []
    title()
    op = input('Enter your Choice:')
    op = int(op)
    if op == 1:
        add_details()
    elif op == 2:
        ano = int(input('Enter account number:'))
        try:
            check_acc_no(ano)
            choices()
        except NameError as ner:
            print('-'*50) 
            print(ner)
            print('-'*50) 
        else:
            print('----------------')
            print('Please give our feedback at 54323')
            print('----------------')
        finally:
            print('----------------')
            print('Thankyou for banking with us !!!')
            print('----------------')
    else:
        break
                   
```
